```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
def running_mean(mylist,N):
    cumsum, moving_aves = [0], []

    for i, x in enumerate(mylist, 1):
        cumsum.append(cumsum[i-1] + x)
        if i>=N:
            moving_ave = (cumsum[i] - cumsum[i-N])/N
            #can do stuff with moving_ave here
            moving_aves.append(moving_ave)
    return moving_aves
```

```python
my_file = open("C:/Users/osatici/Box Sync/4th Semester/CS591/Project/scores.txt", "r")
content_list = my_file.readlines()
content_list = [float(content_list[i][0:-1]) for i in range(len(content_list))]
```

```python
plt.plot(content_list)
plt.title('Episode Scores')
plt.xlabel('episodes')
plt.ylabel('total score')
plt.show()
```

```python

plt.plot(running_mean(content_list,10))
plt.title('Running Average')
plt.xlabel('episodes')
plt.ylabel('total score')
plt.show()
```
